```
import numpy as np
from random import choice
import matplotlib.pyplot as plt
import matplotlib.animation as animation
from IPython.display import HTML
```

```
def isvisited(v):
    if v is not 0:
        return False
    return True

def isallvisited(board):
    for i in range(len(board)):
        for j in range(len(board[i])):
             if not isvisited(board[i,j]):
                    return False
    return True

def get_value(board,coord):
    i,j = coord
    if 0 <= i < len(board):
        if 0 <=  j < len(board[i]):
            return board[i,j]
    return None

def traverse(board,start):
    visit(board,start)


steps = []
c = 0
def visit(board,coord):
    global c
    c += 0.1
    steps.append(coord)
    i,j = coord
    board[i,j] += c
    while not isallvisited(board):
        unvisited = []
        if get_value(board,(i-1,j)) == 0: unvisited.append((i-1,j))
        if get_value(board,(i+1,j)) == 0: unvisited.append((i+1,j))
        if get_value(board,(i,j-1)) == 0: unvisited.append((i,j-1))
        if get_value(board,(i,j+1)) == 0: unvisited.append((i,j+1))
        if len(unvisited) == 0:
            # dead end
            return 
        visit(board,choice(unvisited))
        # backtracking
        steps.append(coord)
```

```
board = np.zeros((10,10))
traverse(board,(0,0))
plt.imshow(board)
```

```
fig, ax = plt.subplots()

c_ = 0
board_ = np.zeros((10,10))
def animate(i):
    global c_
    c_ += 0.1
    board_[steps[i]] = c_
    ax.imshow(board_)

ani = animation.FuncAnimation(fig,animate,interval=100)
plt.close()
```
